```python

df = pd.read_excel('Wie is de Ezel.xlsx', sheet_name="GHS")

print(df)
```

```python
df.columns
```

```python
games = list(df.iterrows())[5:]
wanted_cols = ['Date','Lorne','Max','Marije','Fraser']
```

```python
created_game = Game(name="GHS")
users = {}
created_game.save()

for user in wanted_cols[1:]:
    users[user] = User.objects.get(username=user)
    created_game.users.add(users[user])


created_game.save()
    

for i,game in enumerate(games): 
    
    row = Row(game=created_game)
    
    row.save_no_cols()
    
    
    
    
    for username, user in users.items():
        col = Column(row=row, user=user, value=df[username][5+i])
        col.save()
        
    row.created_at = df['Date'][5+i]
    print(row)
    row.save()
        
```

```python
from scipy.stats.stats import pearsonr
```

```python
scores = {}
for row in Game.objects.get(name='GHS').rows:
    for col in row.columns:
        user_score = scores.get(col.user.username,[])
        user_score.append(col.value)
        scores[col.user.username] = user_score
print(scores)
```

```python
scores_df = pd.DataFrame.from_dict(scores)
```
